Day 1 part 1

```rust
use std::fs;

let mut input = fs::read_to_string("input/day1.txt")?;

let mut cnt_zeros=0;
let mut pos=50;

for line in input.lines(){
    let val: i32 = line[1..].parse().unwrap();
    if line.starts_with("R"){
        pos += val;
    }else{
        pos-= val;
    }
    

    if pos < 0 {pos = 100 + pos % 100;}
    if pos >= 100 {pos = pos % 100;}
    if pos == 0{cnt_zeros+=1;}

}

println!("Num Zeros = {}", cnt_zeros)
```

```rust
let input = fs::read_to_string("input/day1.txt")?;
    let mut cnt_zeros = 0;
    let mut pos = 50;
    
    for line in input.lines() {
        let val: i32 = line[1..].parse().unwrap();
        let init_pos = pos;
        
        if line.starts_with("R") {
            // Moving right (clockwise)
            let distance_to_zero = (100 - init_pos) % 100;
            if distance_to_zero == 0 {
                // Already at 0, count full rotations
                cnt_zeros += val / 100;
            } else if val >= distance_to_zero {
                // Will pass through 0 at least once
                cnt_zeros += 1 + (val - distance_to_zero) / 100;
            }
            pos = (pos + val) % 100;
        } else {
            // Moving left (counterclockwise)
            let distance_to_zero = init_pos;
            if distance_to_zero == 0 {
                // Already at 0, count full rotations
                cnt_zeros += val / 100;
            } else if val >= distance_to_zero {
                // Will pass through 0 at least once
                cnt_zeros += 1 + (val - distance_to_zero) / 100;
            }
            pos = (pos - val % 100 + 100) % 100;
        }
    }
    
    println!("Num Zeros = {}", cnt_zeros);
```
