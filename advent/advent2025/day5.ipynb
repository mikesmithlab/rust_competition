```rust
{
//day5 part1
use std::fs;

let mut input = fs::read_to_string("input/day5.txt")?;

let mut ingredients: Vec<usize> = vec![];
let mut fresh: Vec<Vec<usize>> = vec![];


let mut flag = true;

for line in input.lines(){
    if line == ""{
        flag = false;
    }
    else if flag{
        fresh.push(line.split("-").map(|x| x.parse::<usize>().unwrap()).collect());
    }else {   
        ingredients.push(line.parse::<usize>().unwrap());
    }
}

let mut num_fresh=0;

for ingredient in ingredients.iter(){
    for range in fresh.iter(){
        let start = range[0];
        let stop = range[1];

        if (start..=stop).contains(ingredient){
            num_fresh+=1;
            break;
        }

    }
}



//println!("{:?}", fresh);
//println!("{:?}", ingredients);
println!("num_fresh {:?}", num_fresh);

}
```

```rust
{
//day5 part2
use std::fs;


fn combine_ranges(rng1: Vec<usize>, rng2: &Vec<usize>)-> Option<Vec<usize>>{
    //assumes rng1[0] <= rng2[0]
    let mut new_rng = vec![];

    //
    if rng1[1] < rng2[0]{
        // not overlapping
        return None;
    }else if rng1[1] >= rng2[1]{
        // range 1 encompasses range 2
        new_rng.push(rng1[0]);
        new_rng.push(rng1[1]);
        return Some(new_rng);
    }else {
        //overlapping
        new_rng.push(rng1[0]);
        new_rng.push(rng2[1]);
        return Some(new_rng);
    }

}




let mut input = fs::read_to_string("input/day5.txt")?;

let mut ingredients: Vec<usize> = vec![];
let mut fresh: Vec<Vec<usize>> = vec![];


let mut flag = true;

for line in input.lines(){
    if line == ""{
        break
    }
    else if flag{
        fresh.push(line.split("-").map(|x| x.parse::<usize>().unwrap()).collect());
    }
}

fresh.sort_by(|a,b| a[0].cmp(&b[0]));

let mut new_rng = fresh[0].clone();
let mut total=0;

for i in 0..(fresh.len()-1){
    let output = combine_ranges(new_rng.clone(), &fresh[i+1]);
    println!("{} {:?}", i, output);
    match output{
        Some(val) => {new_rng = val;
                       if i==(fresh.len()-2){
                            total+= (new_rng[1]-new_rng[0] + 1);
                        }                  
                    },
        None => {total += (new_rng[1] - new_rng[0] + 1);
            new_rng = fresh[i+1].clone();
            }
    }
    
}

println!("{:?}", total);

}
```
