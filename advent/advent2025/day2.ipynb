```rust
//day 2 part 1
//In range find len num. Check if first and second half are equal.
//Store ids and sum.
{
    use std::fs;
    let contents = fs::read_to_string("input/day2.txt").unwrap();
    let contents = contents.trim();
    let mut parts: Vec<&str> = contents.split(",").collect();

    println!("{:?}",parts);
    let mut sum_ids = 0;

    for id in parts.iter_mut(){
            let mut ids : Vec<&str>= id.split('-').collect();
            let min_id = ids[0].parse::<usize>().unwrap();
            let max_id = ids[1].parse::<usize>().unwrap();
            //println!("min {} max {}", min_id, max_id);
            for i in min_id..=max_id{
                let len_num = i.to_string().len();
                if len_num%2==0{
                let first = &i.to_string()[..len_num/2];
                let second = &i.to_string()[len_num/2..];
                if first == second {
                    //println!("first {:?} second {:?}", first, second);
                    //println!("id {}",i);
                    sum_ids += i;
                }
            }
            }
    }
    println!("{}",sum_ids);
}
```

```rust
//Day 2 part 2
{
    use std::fs;
    let contents = fs::read_to_string("input/day2.txt").unwrap();
    let contents = contents.trim();
    let mut parts: Vec<&str> = contents.split(",").collect();

    fn check_repeated_n(full_str: &str)-> bool{
        //If any of the length substrings repeats then false_id is true.
        let mut false_id = false;
        for num in 1..=(full_str.len()/2){
            //start of substring
            let test_str = &full_str[..num];
             
            //valid=true
            let mut valid = true;
            let mut j=num;
            if full_str.len()%num==0{
            for i in (0..=(full_str.len()-num)).step_by(num){
                if test_str != &full_str[i..j]{
                    valid = false;
                }
                j+=num;
             }
            }else{
                valid=false;
            }
        
            if valid{
                false_id = true
            }
        }
        
        false_id
    }

    let mut sum_ids = 0;

    for id in parts.iter_mut(){
            let mut ids : Vec<&str>= id.split('-').collect();
            let min_id = ids[0].parse::<usize>().unwrap();
            let max_id = ids[1].parse::<usize>().unwrap();
            for i in min_id..=max_id{
                let full_str = i.to_string();
                
                if check_repeated_n(&full_str) {
                    sum_ids += i;
                }
            
            }
    }
    println!("sum {}",sum_ids);
}
```
