```rust
{
    //day8 part1
    use std::fs;
    use std::collections::HashSet;

    let mut input = fs::read_to_string("input/day8.txt")?;

    //parse coords
    let mut coords: Vec<[i128; 3]> = Vec::with_capacity(1000);

    for line in input.lines(){
        let num_str: Vec<&str> = line.split(",").collect();
        let point : [i128; 3] = [num_str[0].parse::<i128>()?, num_str[1].parse::<i128>()?, num_str[2].parse::<i128>()?];

        coords.push(point); 
    }

    let mut dist : Vec<(i128, usize, usize)> = Vec::with_capacity(1000000);

    for i in 0..coords.len()-1{
        for j in i+1..coords.len(){
            let d = (coords[i][0]-coords[j][0])*(coords[i][0]-coords[j][0]) + (coords[i][1]-coords[j][1])*(coords[i][1]-coords[j][1]) + (coords[i][2]-coords[j][2])*(coords[i][2]-coords[j][2]);
            dist.push((d, i, j));
        }
    }

    dist.sort_by(|a, b| a.0.cmp(&b.0));
    dist.truncate(1000);
    let _len_dist = dist.len();
    

    fn find_clusters(edges: &mut Vec<(i128, usize, usize)>)-> usize{
        //println!("edges {:?}", edges.len());
        
        let mut cluster = HashSet::new();//keep track of items visited

        let (_d, a,b) = edges.swap_remove(0);
        cluster.insert(a);
        cluster.insert(b);

        let mut still_growing = true;

        while still_growing {
            still_growing = false; // Assume we are done unless we find something
            
            let mut i = 0;
            while i < edges.len() {
                let (_d, a, b) = edges[i];

                if cluster.contains(&a) || cluster.contains(&b) {
                    cluster.insert(a);
                    cluster.insert(b);
                    
                    edges.swap_remove(i); 
                    still_growing = true;
                    
                } else {
                    i += 1;
                }
            }
            
        }

        
        //println!("{:?}", cluster.len());
        //println!("{:?}", edges.len());
        cluster.len()
    }
    let mut cluster_lens: Vec<usize> = vec![];

    while dist.len()>0{
        cluster_lens.push(find_clusters(&mut dist));
    }

    cluster_lens.sort_by(|a,b| b.cmp(a));
    //println!("{:?}", cluster_lens);
    println!("{:?}", cluster_lens[0]*cluster_lens[1]*cluster_lens[2]);
}



```
