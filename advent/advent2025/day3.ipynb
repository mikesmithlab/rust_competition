```rust
// day3 part 1
{
    use std::fs;

    fn find_largest_digit(substr: &str)-> char{
        let mut max_val = '0';
        for (index, c) in substr.chars().enumerate(){
            if c > max_val{
                max_val = c;
                if max_val == '9'{
                    break;
                }
            }
        }
        max_val
    }

    let contents = fs::read_to_string("input/day3.txt").unwrap();
    let mut max_val="0".to_string();
    let mut sum_vals = 0;

    for line in contents.lines(){
        //Find first value
        max_val = "0".to_string();
        for i in 0..line.len()-1{
            let sub_str_max = find_largest_digit(&line[i+1..]);
            let test_val = format!("{}{}",&line[i..i+1],sub_str_max);
            
            if test_val > max_val{
                max_val = test_val;               
            }
        }
    sum_vals += max_val.parse::<i32>().unwrap();
         
    }

    println!("Total {}", sum_vals)

    

}
   
```
