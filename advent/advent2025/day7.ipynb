```rust
{
//day7 part1
use std::fs;
use std::collections::HashSet;

let mut input = fs::read_to_string("input/day7.txt")?;



let mut new_tachyon_idx: HashSet<usize> = HashSet::from([input.find("S").unwrap() + 1]);
let mut old_tachyon_idx: HashSet<usize> = HashSet::new();
let mut row: Vec<&str> = vec![];
println!("{:?}", row);
let mut split: usize = 0;

for (i,line) in input.lines().enumerate(){
    //println!("line {}", i);
    
    row = line.split("").collect::<Vec<&str>>();
    //println!("b4    {:?}",row);

    for c in old_tachyon_idx.iter(){
        if row[*c] == "."{
            new_tachyon_idx.insert(*c);
            row[*c] = "|";

        }
        else if row[*c] == "^"{
            new_tachyon_idx.insert(*c-1);
            row[*c-1] = "|";
            new_tachyon_idx.insert(*c+1);
            row[*c+1] = "|";
            split+=1;
        }
        
    }

    //println!("after {:?}", row);

    //println!("{:?}", old_tachyon_idx);
    //println!("{:?}", new_tachyon_idx);
    
    

    old_tachyon_idx = new_tachyon_idx.clone();
    new_tachyon_idx.clear();

}

println!("{:?}",split);


}
```

```rust
{
//day7 part2
use std::fs;
use std::collections::HashMap;

let mut input = fs::read_to_string("input/day7.txt")?;



// <index, num_paths>
let mut new_tachyon_idx: HashMap<usize, usize> = HashMap::new();
let mut old_tachyon_idx: HashMap<usize, usize> = HashMap::new();
new_tachyon_idx.insert(input.find("S").unwrap()+1, 1);

let mut row: Vec<&str> = vec![];
println!("{:?}", row);
let mut split: usize = 0;

for (i,line) in input.lines().enumerate(){
    //println!("line {}", i);
    
    row = line.split("").collect::<Vec<&str>>();
    //println!("b4    {:?}",row);

    for (idx, num) in old_tachyon_idx.iter(){
        if row[*idx] == "."{
            *new_tachyon_idx.entry(*idx).or_insert(0) += num;
            row[*idx] = "|";

        }
        else if row[*idx] == "^"{
            *new_tachyon_idx.entry(*idx-1).or_insert(0) += num;
            //row[*idx-1] = "|";
            *new_tachyon_idx.entry(*idx+1).or_insert(0) += num;
            //row[*idx+1] = "|";
            split+=1;
        }
        
    }

    //println!("after {:?}", row);

    //println!("{:?}", old_tachyon_idx);
    //println!("{:?}", new_tachyon_idx);
    
    

    old_tachyon_idx = new_tachyon_idx.clone();
    new_tachyon_idx.clear();

}

let mut total = 0;
for val in old_tachyon_idx.values(){
    total += val;
}

println!("{:?}",total);


}
```
