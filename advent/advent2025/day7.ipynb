```rust
{
//day6 part1
use std::fs;

let mut input = fs::read_to_string("input/day6.txt")?;

let lines: Vec<&str> = input.split("\n").collect();

let num_lines = lines.len() - 1; //last line is blank
let len_lines = lines[0].len();
let mut nums: Vec<Vec<&str>> = vec![];
let mut symbols: Vec<&str>= vec![];

for (i,line) in lines.iter().enumerate(){
    if i < (num_lines-1){//last line is blank
        nums.push(line.split_whitespace().collect());
    }else{
        symbols = line.split_whitespace().collect();
        break;
    }
  
}


let mut total: u64 = 0;
let mut subtotal: u64 = 0;


for i in 0..symbols.len(){
    subtotal=0;
    for j in 0..(num_lines-1){
            if symbols[i] == "*"{
                if subtotal == 0{
                    subtotal = 1;
                }
                subtotal *= nums[j][i].parse::<u64>().unwrap();
            }else{
                subtotal += nums[j][i].parse::<u64>().unwrap();
            }

        }

    total+= subtotal;

    }

 

println!("total = {}", total);

}
```

```rust
{
//day6 part2
use std::fs;

let mut input = fs::read_to_string("input/day6.txt")?;

let lines: Vec<&str> = input.split("\n").collect();

let num_lines = lines.len() - 1; //last line is blank
let len_lines = lines[0].len();
let mut nums: Vec<Vec<&str>> = vec![];
let mut symbols: Vec<&str>= vec![];

for (i,line) in lines.iter().enumerate(){
    if i < (num_lines-1){//last line is blank
        let mut new_line: Vec<&str> = line.split("").collect();
        new_line.pop();
        new_line.pop();
        new_line.remove(0);
        nums.push(new_line);
    }else{
        symbols = line.split_whitespace().collect();
        break;
    }
  
}

//clean up the lines


println!("{:?}", nums[0]);
println!("{:?}", nums[1]);
println!("{:?}", nums[2]);
println!("{:?}", symbols);


let mut total: u64 = 0;
let mut subtotal: u64 = 1;
let mut value: usize = 0;
let mut num: String = "".to_string();
let mut current_symbol = "*";

let mut sym_idx=0;
    //println!("New column");
    //println!("subtotal = {}", subtotal);    
    for i in 0..=nums[0].len(){
        //println!("i {}", i);
        value = 0;
        num = "".to_string();
        if i == nums[0].len(){
                total+=subtotal;
                break;
            }
        for j in 0..(num_lines-1){
            //println!("j {}, nums[j][i] {}, num {}", j, nums[j][i], num);
            //Test if this column is a number
            
            //Test for empty column
            if (nums[j][i] == " ") {
                //Don't do any operations but log it is empty
                //When get whole column update symbol and move on
                value += 1;
                if value == (num_lines-1){
                    //println!("subtotal to be added {}", subtotal);
                    total+= subtotal;
                    //println!("total {}", total);
                    //println!("Empty column");
                    sym_idx += 1;
                    current_symbol = symbols[sym_idx];
                    if current_symbol == "*"{
                        subtotal = 1;
                    } else if current_symbol == "+"{
                        subtotal = 0;
                    }
                    //println!("current symbol {}", current_symbol);
                    break;
                }

            }
            //If digit is not empty space concatenate digit to string
            if nums[j][i] != " "{
                num = num + nums[j][i];
                //println!("i {} j {} nums[j][i] {} num {}", i, j, nums[j][i], num)
            }
            //println!("num {}", num);
            //If this is the last row
            if j == num_lines-2{
                if current_symbol == "*"{
                subtotal *= num.parse::<u64>().unwrap();
                }
                else if current_symbol == "+"{
                    subtotal += num.parse::<u64>().unwrap();
                }
            //println!("subtotal {}", subtotal);
            }
        }
    }
    //println!("subtotal = {}, total = {}, current_symbol {}", subtotal, total, current_symbol );

    

    println!("total = {}", total);

}

```
