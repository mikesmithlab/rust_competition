```rust
//Day 4 part 2
{
use std::fs;

fn sum_subgrid(grid: &Vec<Vec<i32>>, coord: &(usize, usize))->i32{
    let mut i=coord.0;
    let mut j=coord.1;

    let mut sum_val = 0;
    for r in i-1..=i+1{
        for c in j-1..=j+1{
            if r!=i || c!=j{
                //println!("r{}c{} val{}", r,c,grid[r][c]);
                sum_val += grid[r][c];
            }
        }
    }
    //println!("sum_val {}, coord {:?}", sum_val, coord);

    if sum_val < 4{
        1
    }
    else {
     0   
    }
    
}

fn update_grid(grid: Vec<Vec<i32>>, coords: &Vec<(usize,usize)>) -> Vec<Vec<i32>>
{
    let width = grid.len();
    let height = grid[0].len();
    //returns grid with just coords
    let mut new_grid: Vec<Vec<i32>> = vec![vec![0; width]; height];

    for coord in coords.iter(){
        new_grid[coord.0][coord.1] = 1;
    }

    new_grid


}


let mut input = fs::read_to_string("input/day4.txt")?;
//let mut input = "@@@\n@@@\n@@@";

let mut grid = vec![];
let mut coords = vec![];

let mut grid_width=0;



for (i, line) in input.lines().enumerate(){
    //add a line of 0s to the top
    if i == 0{
        grid_width = line.len();
        let row_zeros = vec![0;grid_width+2];
        grid.push(row_zeros);
        
    }
    
    //Create grid from file
    let mut row = vec![];
    for (j, c) in line.chars().enumerate(){
        if j==0{
        row.push(0);
        }
        if c == '@'{
            row.push(1);
            coords.push((i+1,j+1));
        }else {
         row.push(0)   
        }
    }
    row.push(0);
    grid.push(row);   
}
grid.push(vec![0;grid_width+2]);


println!("0 start grid {:?}", grid);
println!("0 start coords {:?}",coords);

let mut total = 0;
let mut subtotal=0;
let mut i=0;
    loop{
        
        let mut remove_coords = vec![];
        subtotal = 0;
        //println!("subtotal reset {}", subtotal);
        //println!("coords {:?}", coords);
        //println!("grid {:?}", grid);
        for coord in coords.iter(){
            let sum_val = sum_subgrid(&grid, coord);
            if sum_val == 1{
                remove_coords.push(coord);
                subtotal+=sum_val;
            }
            total+=sum_val;
            
        }
        //println!("subtotal after loop {}", subtotal);
        //println!("{} remove_coords {:?}", i, remove_coords);
        coords = coords.iter().filter(|item| !remove_coords.contains(item)).copied().collect();
        //println!("{} new coords {:?}", i, coords);
        grid =  update_grid(grid.clone(), &coords);
        //println!("{} new grid {:?}", i, grid);

        //println!("subtotal {}", subtotal);
        if subtotal == 0{
            break;
        }
    
    }

println!("{:?}", grid);

println!("{}", total);

}
```
